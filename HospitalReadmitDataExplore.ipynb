```python
def get_cut_off_exponnorm(data, alpha):
    K, loc, scale = ss.exponnorm.fit(data)  # get K value 
    cut_off = ss.exponnorm.ppf(q = 1.0 - alpha, K=K, loc = 0, scale = 1)
    return cut_off
```

```python
ami_df_top_cut_off = get_cut_off_exponnorm(ami_df.actual_readmit_rate, .95)
ami_df_top_cut_off
```

```python
def get_top_performance_hospitals(df, col_readmit_rate, critical_pt):
    cut_off = df[col_readmit_rate].mean() - critical_pt * df[col_readmit_rate].std()
    top_df = df[df[col_readmit_rate] < cut_off]
    return top_df
```

```python
def get_poor_performance_hospitals(df, col_readmit_rate, critical_pt):
    cut_off = df[col_readmit_rate].mean() + critical_pt * df[col_readmit_rate].std()
    poor_df = df[df[col_readmit_rate] > cut_off]
    return poor_df
```

Thought about using Wilcoxon one sample test and Wilcoxon Signed Rank Test and use median instead of mean for defining readmission rate 'averages'.  The results are not really any different so I tabled this for another time.

```python
#confidence interval using wilcoxon one sample test, conf. interval is from the rth to the sth ranked item
#r = n/2 - (1.96 * n**0.5)/2
#s = 1 + n/2 + (1.96 * sqrt(n))/2
import math
def get_median_ci(data):
    n = len(data)
    median = data.median()
    sorted_data = data.sort_values().reset_index()
    del sorted_data['index']
    r = int(n/2 - 1.96 * math.sqrt(n)/2)
    s = int(1 + n/2 + 1.96 * math.sqrt(n)/2)
    lower_ci_value = sorted_data.loc[r-1]
    upper_ci_value = sorted_data.loc[s-1]
    return median, lower_ci_value, upper_ci_value
```
